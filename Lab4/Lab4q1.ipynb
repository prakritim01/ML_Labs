```python
import random
import matplotlib.pyplot as plt
```

```python
# Utility functions 
def compute_mean(data):
    return sum(data) / len(data)

def compute_variance(data):
    mean = compute_mean(data)
    return sum((x - mean) ** 2 for x in data) / len(data)
```

```python
# Dice rolling experiment
def roll_dice(num_rolls):
    """Simulate rolling a fair 6-sided die num_rolls times and return the sum."""
    return sum(random.randint(1, 6) for _ in range(num_rolls))

def dice_experiment(num_rolls, num_experiments):
    """Perform num_experiments of rolling a die num_rolls times, return sums list."""
    results = [roll_dice(num_rolls) for _ in range(num_experiments)]
    return results
```

```python
# Plotting function 
def plot_histogram(data, title, bins=20):
    plt.hist(data, bins=bins, edgecolor='black', alpha=0.7)
    plt.title(title)
    plt.xlabel("Sum of Rolls")
    plt.ylabel("Frequency")
    plt.show()
```

```python
# Main Experiment
num_experiments_list = [100, 500, 1000, 5000, 10000, 50000, 100000]

for rolls in [1, 2, 3]:  
    for n_exp in num_experiments_list:
        results = dice_experiment(rolls, n_exp)
        mean_val = compute_mean(results)
        var_val = compute_variance(results)
        print(f"Rolls: {rolls}, Experiments: {n_exp}, Mean: {mean_val:.2f}, Variance: {var_val:.2f}")
        plot_histogram(results, f"Histogram of {rolls} Rolls ({n_exp} Experiments)")
```
